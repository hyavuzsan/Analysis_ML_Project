```python
# Dosya yolu
file_path = "Data/NetAttendence.xlsx"
sheets = ["One year before", "Primary", "Lower secondary", "Upper secondary"]

# Veri setini yükleme
data_frames = []
for sheet in sheets:
    df = pd.read_excel(file_path, sheet_name=sheet)
    df["Education Level"] = sheet
    data_frames.append(df)
data = pd.concat(data_frames, ignore_index=True)

# İlk 10 satırı göster
data.head(10)
```

```python
# Boşluk içeren sütun isimlerini düzelt
data.columns = [str(col).replace(" ", "") for col in data.columns]

# Unnamed sütunları kaldır
unnamed_cols = [col for col in data.columns if "Unnamed" in str(col)]
data.drop(columns=unnamed_cols, inplace=True, errors="ignore")

# Sütunları yeniden adlandırma
data.rename(columns={
    "Country": "Country",
    "Region": "Region",
    "EducationLevel": "Education Level",
    "Total": "Value",
}, inplace=True)

# Sayısal sütun dönüşümleri
data["Value"] = pd.to_numeric(data["Value"], errors="coerce")
data.dropna(subset=["Value"], inplace=True)

# Temizlenmiş ilk 10 satır
data.head(10)
```

Value sütunu ülkenin eğitime katılan nüfusunun yüzdesini gösteriyor.

```python
import matplotlib.pyplot as plt

# Kıtaların İngilizce kısaltmalarını Türkçe isimlerle değiştirme
region_mapping = {
    "EAP": "Doğu Asya ve Pasifik",
    "ECA": "Avrupa ve Orta Asya",
    "MENA": "Orta Doğu ve Kuzey Afrika",
    "SSA": "Sahra Altı Afrika",
    "LAC": "Latin Amerika ve Karayipler",
    "SA": "Güney Asya"
}

# Kıtaların isimlerini değiştirme ve eksik değerleri kontrol etme
data["Region"] = data["Region"].map(region_mapping)
data = data.dropna(subset=["Region"])  

# Eğitime katılım oranlarını gruplandırma
mean_values = data.groupby(["Region", "Education Level"])["Value"].mean().unstack()

# Bar grafiği oluştur
ax = mean_values.plot(kind="bar", figsize=(14, 8), edgecolor="black", width=0.8)

# Ekseni etiketleme
ax.set_xlabel("Kıtalar")
ax.set_ylabel("Katılım Oranı (%)")
ax.set_title("Kıtalar Bazında Eğitime Katılım Oranları (Ortalama)")


plt.xticks(ticks=range(len(mean_values.index)), labels=mean_values.index, rotation=45)

# Legend (Açıklama) ve Grid ekleme
plt.legend(title="Eğitim Seviyesi")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Grafik alt boşluğu düzenleme
plt.subplots_adjust(bottom=0.25)

# Grafiği gösterme
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


continent_names = [
    "East Asia & Pacific", "Europe & Central Asia", "Latin America & Caribbean",
    "Middle East & North Africa", "South Asia", "Sub-Saharan Africa"
]

# Her bölge için analiz yapalım
regions = data["Region"].unique()

for region in regions:
    region_data = data[data["Region"] == region]

    # Eğer bölgede veri yoksa atla
    if region_data.empty:
        print(f"⚠ {region} bölgesi için veri bulunamadı!")
        continue

    # Kıta isimlerini veriden çıkaralım (sadece ülkeler kalsın)
    region_data = region_data[~region_data["Country"].isin(continent_names)]

    # Eksik verileri ülke bazında ortalama ile dolduralım
    region_data["Value"] = region_data.groupby("Country")["Value"].transform(lambda x: x.fillna(x.mean()))

    # Ülkelere göre eğitime katılım oranlarını hesaplayalım
    mean_values = region_data.groupby("Country")["Value"].mean().sort_values(ascending=False)

    # Eğer ülkelerde hiç veri yoksa atla
    if mean_values.empty:
        print(f"⚠ {region} bölgesi için veri eksik, grafik oluşturulamadı!")
        continue

    # GRAFİK OLUŞTURMA  
    plt.figure(figsize=(14, 7))
    mean_values.plot(kind="bar", color="coral", edgecolor="black")

    # Grafik Başlığı ve Eksen Etiketleri
    plt.title(f"{region.lower()} Bölgesindeki Ülkelerin Katılım Oranı", fontsize=16)
    plt.xlabel("Ülkeler", fontsize=12)
    plt.ylabel("Katılım Oranı (%)", fontsize=12)

    #  X eksenindeki ülke isimlerini düzenleme
    plt.xticks(rotation=45, ha="right", fontsize=10)

    #  Izgara Çizgileri
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # Kenarları düzgün göster
    plt.tight_layout()

    #  Grafiği göster
    plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Veri Setlerini Yükle
file_path_attendance = r"Data\NetAttendence.xlsx"
file_path_internet = r"Data\DigitalConnectivity.xlsx"

# Eğitime Katılım Verisini Okuma
sheets = ["One year before", "Primary", "Lower secondary", "Upper secondary"]
attendance_data = []
for sheet in sheets:
    df_att = pd.read_excel(file_path_attendance, sheet_name=sheet)
    df_att["Education Level"] = sheet  # Eğitim seviyesini belirtiyoruz
    attendance_data.append(df_att)

# Tüm eğitim seviyelerini birleştir
attendance_df = pd.concat(attendance_data, ignore_index=True)

# Sütunları düzenle
attendance_df.rename(columns={
    "Country": "Country",
    "Total": "Attendance Rate"
}, inplace=True)

attendance_df = attendance_df[["Country", "Attendance Rate"]]

# Eksik verileri temizle
attendance_df.dropna(subset=["Attendance Rate"], inplace=True)

# İnternete Erişim Verisini Okuma
internet_df = pd.read_excel(file_path_internet, sheet_name="Upper Secondary")

# Sütunları düzenle
internet_df.rename(columns={
    "Countries and areas": "Country",
    "Total": "Internet Access"
}, inplace=True)

internet_df = internet_df[["Country", "Internet Access"]]

# Yüzde işaretini kaldır ve numerik hale getir
internet_df["Internet Access"] = internet_df["Internet Access"].astype(str).str.replace('%', '').str.replace(',', '.')
internet_df["Internet Access"] = pd.to_numeric(internet_df["Internet Access"], errors='coerce')

attendance_df["Attendance Rate"] = pd.to_numeric(attendance_df["Attendance Rate"], errors='coerce')

# Eksik verileri temizle
attendance_df.dropna(subset=["Attendance Rate"], inplace=True)
internet_df.dropna(subset=["Internet Access"], inplace=True)

# Veri Setlerini Birleştir
merged_df = pd.merge(attendance_df, internet_df, on="Country")

# Pearson Korelasyonunu Hesapla
correlation, p_value = pearsonr(merged_df["Internet Access"], merged_df["Attendance Rate"])

print(f"📊 Pearson Korelasyon Katsayısı: {correlation:.2f}")
print(f"📌 p-value: {p_value:.5f}")

# Scatter Plot (Dağılım Grafiği) Çiz
# Scatter Plot + Regresyon Çizgisi
plt.figure(figsize=(10, 6))
sns.regplot(x="Internet Access", y="Attendance Rate", data=merged_df, scatter_kws={"alpha":0.6}, line_kws={"color":"red"})
plt.xlabel("İnternete Erişim Oranı (%)")
plt.ylabel("Eğitime Katılım Oranı (%)")
plt.title("İnternete Erişim ile Eğitime Katılım Arasındaki İlişki (Regresyon Çizgili)")
plt.grid(True)
plt.show()



```

Korelasyon analizi yaparak aşağıdaki çıkarımlarda bulunduk.

📌 Grafik Yorumu:

Regresyon çizgisi (kırmızı çizgi) yukarı eğimli → İnternete erişim arttıkça eğitime katılım da artıyor.
Pek çok veri noktası kırmızı çizgiye yakın, ancak bir kısmı dağınık → Orta derecede pozitif korelasyon (0.53) olduğunu destekliyor.
Alt kısımda (0-20% internet erişimi) eğitime katılım oranı düşük → İnternete erişimi olmayan ülkelerde eğitim seviyesi de düşük.
Üst kısımda (80%+ internet erişimi) eğitime katılım oranı genelde yüksek → Dijital altyapı iyi olan ülkeler eğitime daha fazla erişebiliyor.

Eğitime katılımı etkileyen diğer faktörler de olabilir (örneğin gelir seviyesi, hükümet politikaları, kültürel engeller).

```python

        # CmdStanpy ve FutureWarning mesajlarını kapat
        logging.getLogger("cmdstanpy").setLevel(logging.ERROR)
        warnings.simplefilter(action='ignore', category=FutureWarning)

        file_path = r"Data\Data.csv"

        df = pd.read_csv(file_path, encoding="ISO-8859-1")

        # Seçilen göstergeyi filtreliyoruz
        selected_indicator = "Completion rate, primary education, both sexes (%)"
        df_filtered = df[df["Series"] == selected_indicator]

        # Gereksiz sütunları kaldırıyoruz ve yıllık verileri uzun formata çevir
        df_filtered = df_filtered.drop(columns=["Series", "Series Code"])
        df_filtered = df_filtered.melt(id_vars=["Country Name", "Country Code"],
                                       var_name="Year",
                                       value_name="Completion Rate")

        # Yıl sütununu temizleyelim ve tam tarih formatına çevirelim
        df_filtered["Year"] = df_filtered["Year"].str.extract("(\\d{4})").astype(int)
        df_filtered = df_filtered[df_filtered["Year"] <= 2020]  # 2020'ye kadar olan verileri alalım
        df_filtered["Year"] = pd.to_datetime(df_filtered["Year"].astype(str) + "-01-01")

        # Completion Rate sütunundaki eksik değerleri yönetelim
        df_filtered["Completion Rate"] = pd.to_numeric(df_filtered["Completion Rate"], errors="coerce")

        # Ülke bazında eksik veri yüzdesini hesaplayalım
        missing_percentage = df_filtered.groupby("Country Name")["Completion Rate"].apply(lambda x: x.isnull().mean())

        # %50'den fazla eksik verisi olan ülkeleri silelim
        countries_to_drop = missing_percentage[missing_percentage > 0.5].index
        df_filtered = df_filtered[~df_filtered["Country Name"].isin(countries_to_drop)]

        # Geri kalan eksik verileri ülke ortalamasıyla dolduralım
        df_filtered["Completion Rate"] = df_filtered.groupby("Country Name")["Completion Rate"].transform(
            lambda x: x.fillna(x.mean()))

        # Daha fazla yıl göstermek için ilk 10 satırı alalım
        df_head = df_filtered.head(10)

        # İlk 10 satırı yazdır
        print(df_head.to_string(index=False))

        # Global analiz için tüm ülkelerin ortalamasını alalım
        df_global = df_filtered.groupby("Year")["Completion Rate"].mean().reset_index()

        # 2020'deki düşüşü düzeltmek için 2021 sonrası için bir toparlanma trendi ekleyelim
        df_global.loc[df_global["Year"] == pd.Timestamp("2020-01-01"), "Completion Rate"] = \
        df_global[df_global["Year"] < pd.Timestamp("2020-01-01")]["Completion Rate"].mean()

        # Prophet için veri çerçevesini uygun hale getirelim
        df_prophet = df_global.rename(columns={"Year": "ds", "Completion Rate": "y"})

        # Prophet modelini oluşturuyoruz
        model = Prophet(
            changepoint_prior_scale=0.08,
            seasonality_mode="additive",
            yearly_seasonality=True
        )

        # Manuel değişim noktaları ekleyelim (Covid-19 etkisini azaltacak şekilde)
        model.add_seasonality(name='yearly', period=1, fourier_order=10)  # Yıllık etki ekleyelim
        model.changepoints = df_prophet[(df_prophet['ds'] >= '2005-01-01') & (df_prophet['ds'] <= '2020-01-01')]['ds']

        # Modeli eğitelim
        model.fit(df_prophet)

        # Gelecek 5 yıl için tahmin yapalım (2021-2025)
        future = model.make_future_dataframe(periods=5, freq='Y')
        future['cap'] = 100  # Tahminin üst sınırını belirleyelim
        future['floor'] = df_prophet['y'].min()
        forecast = model.predict(future)

        # Görselleştirme
        plt.figure(figsize=(12, 6))
        plt.plot(df_prophet["ds"], df_prophet["y"], label="Gerçek Değerler", marker="o", color="blue")
        plt.plot(forecast["ds"], forecast["yhat"], label="Tahmin (Prophet)", linestyle="dashed", marker="o",
                 color="red")
        plt.fill_between(forecast["ds"], forecast["yhat_lower"], forecast["yhat_upper"], alpha=0.1,
                         label="Güven Aralığı")
        plt.xlabel("Yıl")
        plt.ylabel("Tamamlama Oranı (%)")
        plt.title("Global İlköğretim Tamamlama Oranı Tahmini (Prophet) (2021-2025)")
        plt.legend()
        plt.grid(True)
        plt.show()
```

Veri seti ülkelerin ilköğretimi tamamlama oranlarını gösteriyor.

→ 2000-2019 arasında istikrarlı bir artış gözlemlendi ve oran %95 seviyelerine ulaştı.

```python
import pandas as pd

# Dosya Yolunu Belirleme
file_path = r"Data\share-of-the-world-population-with-at-least-basic-education.csv"

# Veri Setini Okuma
df = pd.read_csv(file_path, encoding="ISO-8859-1")

#  Gerekli Sütunları Seçme ve Temizlik
df = df.rename(columns={"Entity": "Country", "Year": "Year",
                        "Share of population with no formal education, 1820-2020": "No_Education_Share"})

#  Sadece 2020 Yılı Verisini Filtreleme
df = df[df["Year"] == 2020]

#  Gerekli Sütunları Seç ve Eksik Değerleri Temizle
df = df[["Country", "Year", "No_Education_Share"]].dropna()

#  Son 10 Satırı Yazdır
print(df.tail(10).to_string(index=False))


```

```python
from sklearn.cluster import KMeans

# K-Means Kümeleme (3 Küme)
X = df[["No_Education_Share"]].values
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(X)

# KÜMELEME SONUÇLARINI CSV'YE KAYDET
df.to_csv("clustering_results.csv", index=False)

print("Kümeleme başarıyla tamamlandı!")
```

```python
# Küme 1 İçindeki Ülkeleri Seç
df_cluster_1 = df[df["Cluster"] == 1].copy()

# Küme 1’i 4 Alt Kümeye Ayır
X_sub = df_cluster_1[["No_Education_Share"]].values
kmeans_sub = KMeans(n_clusters=4, random_state=42, n_init=10)
df_cluster_1["Sub_Cluster"] = kmeans_sub.fit_predict(X_sub)

# YENİ KÜMELERİ CSV’YE KAYDET
df_cluster_1.to_csv("cluster_1_subgroups.csv", index=False)

print(" Küme 1, 4 alt kümeye bölündü!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# NOKTA GRAFİĞİ
plt.figure(figsize=(14, 8))
sns.stripplot(x="Cluster", y="No_Education_Share", data=df, hue="Cluster", jitter=True, palette="Set1",
              alpha=0.7, legend=False)
plt.xlabel("Küme")
plt.ylabel("No Education Share (%)")
plt.title("Genel Eğitim Durumu Nokta Grafiği")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

📌 Genel Eğitim Durumu Nokta Grafiği

🔹 Küme 0 (Kırmızı): Eğitim almayan nüfus oranı orta seviyede (%10-30).
🔹 Küme 1 (Mavi): Eğitim almayan nüfus oranı en düşük seviyede (0-10%).
🔹 Küme 2 (Yeşil): Eğitim almayan nüfus oranı çok yüksek (%40-70), yani bu kümede eğitim erişimi ciddi şekilde kısıtlı.
