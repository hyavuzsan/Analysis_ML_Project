```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
import sys
import logging
import warnings
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


sys.stderr = open('nul', 'w')

# CmdStanpy ve Prophet uyarƒ±larƒ±nƒ± kapatƒ±yoruz
logging.getLogger("cmdstanpy").setLevel(logging.ERROR)
logging.getLogger("prophet").setLevel(logging.ERROR)


# FutureWarning mesajlarƒ±nƒ± kapatƒ±yoruz
warnings.simplefilter(action='ignore', category=FutureWarning)

```

```python
# Dosya yolu
file_path = "Data/NetAttendence.xlsx"
sheets = ["One year before", "Primary", "Lower secondary", "Upper secondary"]

# Veri setini y√ºkleme
data_frames = []
for sheet in sheets:
    df = pd.read_excel(file_path, sheet_name=sheet)
    df["Education Level"] = sheet
    data_frames.append(df)
data = pd.concat(data_frames, ignore_index=True)

# ƒ∞lk 10 satƒ±rƒ± g√∂ster
data.head(10)
```

```python
# Bo≈üluk i√ßeren s√ºtun isimlerini d√ºzelt
data.columns = [str(col).replace(" ", "") for col in data.columns]

# Unnamed s√ºtunlarƒ± kaldƒ±r
unnamed_cols = [col for col in data.columns if "Unnamed" in str(col)]
data.drop(columns=unnamed_cols, inplace=True, errors="ignore")

# S√ºtunlarƒ± yeniden adlandƒ±rma
data.rename(columns={
    "Country": "Country",
    "Region": "Region",
    "EducationLevel": "Education Level",
    "Total": "Value",
}, inplace=True)

# Sayƒ±sal s√ºtun d√∂n√º≈ü√ºmleri
data["Value"] = pd.to_numeric(data["Value"], errors="coerce")
data.dropna(subset=["Value"], inplace=True)

# Temizlenmi≈ü ilk 10 satƒ±r
data.head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Veri Setlerini Y√ºkle
file_path_attendance = r"Data\NetAttendence.xlsx"
file_path_internet = r"Data\DigitalConnectivity.xlsx"

# Eƒüitime Katƒ±lƒ±m Verisini Okuma
sheets = ["One year before", "Primary", "Lower secondary", "Upper secondary"]
attendance_data = []
for sheet in sheets:
    df_att = pd.read_excel(file_path_attendance, sheet_name=sheet)
    df_att["Education Level"] = sheet  # Eƒüitim seviyesini belirtiyoruz
    attendance_data.append(df_att)

# T√ºm eƒüitim seviyelerini birle≈ütir
attendance_df = pd.concat(attendance_data, ignore_index=True)

# S√ºtunlarƒ± d√ºzenle
attendance_df.rename(columns={
    "Country": "Country",
    "Total": "Attendance Rate"
}, inplace=True)

attendance_df = attendance_df[["Country", "Attendance Rate"]]

# Eksik verileri temizle
attendance_df.dropna(subset=["Attendance Rate"], inplace=True)

# ƒ∞nternete Eri≈üim Verisini Okuma
internet_df = pd.read_excel(file_path_internet, sheet_name="Upper Secondary")

# S√ºtunlarƒ± d√ºzenle
internet_df.rename(columns={
    "Countries and areas": "Country",
    "Total": "Internet Access"
}, inplace=True)

internet_df = internet_df[["Country", "Internet Access"]]

# Y√ºzde i≈üaretini kaldƒ±r ve numerik hale getir
internet_df["Internet Access"] = internet_df["Internet Access"].astype(str).str.replace('%', '').str.replace(',', '.')
internet_df["Internet Access"] = pd.to_numeric(internet_df["Internet Access"], errors='coerce')

attendance_df["Attendance Rate"] = pd.to_numeric(attendance_df["Attendance Rate"], errors='coerce')

# Eksik verileri temizle
attendance_df.dropna(subset=["Attendance Rate"], inplace=True)
internet_df.dropna(subset=["Internet Access"], inplace=True)

# Veri Setlerini Birle≈ütir
merged_df = pd.merge(attendance_df, internet_df, on="Country")

# Pearson Korelasyonunu Hesapla
correlation, p_value = pearsonr(merged_df["Internet Access"], merged_df["Attendance Rate"])

print(f"üìä Pearson Korelasyon Katsayƒ±sƒ±: {correlation:.2f}")
print(f"üìå p-value: {p_value:.5f}")

# Scatter Plot (Daƒüƒ±lƒ±m Grafiƒüi) √áiz
# Scatter Plot + Regresyon √áizgisi
plt.figure(figsize=(10, 6))
sns.regplot(x="Internet Access", y="Attendance Rate", data=merged_df, scatter_kws={"alpha":0.6}, line_kws={"color":"red"})
plt.xlabel("ƒ∞nternete Eri≈üim Oranƒ± (%)")
plt.ylabel("Eƒüitime Katƒ±lƒ±m Oranƒ± (%)")
plt.title("ƒ∞nternete Eri≈üim ile Eƒüitime Katƒ±lƒ±m Arasƒ±ndaki ƒ∞li≈üki (Regresyon √áizgili)")
plt.grid(True)
plt.show()



```

Korelasyon analizi yaparak a≈üaƒüƒ±daki √ßƒ±karƒ±mlarda bulunduk.

üìå Grafik Yorumu:

Pek √ßok veri noktasƒ± kƒ±rmƒ±zƒ± √ßizgiye yakƒ±n, ancak bir kƒ±smƒ± daƒüƒ±nƒ±k ‚Üí Orta derecede pozitif korelasyon (0.53) olduƒüunu destekliyor.

√úst kƒ±sƒ±mda (80%+ internet eri≈üimi) eƒüitime katƒ±lƒ±m oranƒ± genelde y√ºksek ‚Üí Dijital altyapƒ± iyi olan √ºlkeler eƒüitime daha fazla eri≈üebiliyor.

Eƒüitime katƒ±lƒ±mƒ± etkileyen diƒüer fakt√∂rler de olabilir (√∂rneƒüin gelir seviyesi, h√ºk√ºmet politikalarƒ±, k√ºlt√ºrel engeller).

‚Üí Ancak Haiti ve Guatemala gibi √ºlkelerde daha d√º≈ü√ºk seviyeler var.

üìå Avrupa ve Orta Asya (ECA)
‚Üí Katƒ±lƒ±m oranlarƒ± olduk√ßa y√ºksek ve √ßoƒüu √ºlkede istikrarlƒ±.

üìå G√ºney Asya (SA)
‚Üí Katƒ±lƒ±m oranƒ± √ºlkeden √ºlkeye b√ºy√ºk farklƒ±lƒ±k g√∂steriyor.

‚Üí Geli≈ümi≈ü √ºlkeler (Japonya, √áin) √ßok y√ºksek, bazƒ± ada √ºlkeleri ise d√º≈ü√ºk seviyede.

üìå Orta Doƒüu ve Kuzey Afrika (MENA)
‚Üí Eƒüitime katƒ±lƒ±m b√∂lge i√ßinde b√ºy√ºk farklƒ±lƒ±k g√∂steriyor.

```python

        # CmdStanpy ve FutureWarning mesajlarƒ±nƒ± kapat
        logging.getLogger("cmdstanpy").setLevel(logging.ERROR)
        warnings.simplefilter(action='ignore', category=FutureWarning)

        file_path = r"Data\Data.csv"

        df = pd.read_csv(file_path, encoding="ISO-8859-1")

        # Se√ßilen g√∂stergeyi filtreliyoruz
        selected_indicator = "Completion rate, primary education, both sexes (%)"
        df_filtered = df[df["Series"] == selected_indicator]

        # Gereksiz s√ºtunlarƒ± kaldƒ±rƒ±yoruz ve yƒ±llƒ±k verileri uzun formata √ßevir
        df_filtered = df_filtered.drop(columns=["Series", "Series Code"])
        df_filtered = df_filtered.melt(id_vars=["Country Name", "Country Code"],
                                       var_name="Year",
                                       value_name="Completion Rate")

        # Yƒ±l s√ºtununu temizleyelim ve tam tarih formatƒ±na √ßevirelim
        df_filtered["Year"] = df_filtered["Year"].str.extract("(\\d{4})").astype(int)
        df_filtered = df_filtered[df_filtered["Year"] <= 2020]  # 2020'ye kadar olan verileri alalƒ±m
        df_filtered["Year"] = pd.to_datetime(df_filtered["Year"].astype(str) + "-01-01")

        # Completion Rate s√ºtunundaki eksik deƒüerleri y√∂netelim
        df_filtered["Completion Rate"] = pd.to_numeric(df_filtered["Completion Rate"], errors="coerce")

        # √úlke bazƒ±nda eksik veri y√ºzdesini hesaplayalƒ±m
        missing_percentage = df_filtered.groupby("Country Name")["Completion Rate"].apply(lambda x: x.isnull().mean())

        # %50'den fazla eksik verisi olan √ºlkeleri silelim
        countries_to_drop = missing_percentage[missing_percentage > 0.5].index
        df_filtered = df_filtered[~df_filtered["Country Name"].isin(countries_to_drop)]

        # Geri kalan eksik verileri √ºlke ortalamasƒ±yla dolduralƒ±m
        df_filtered["Completion Rate"] = df_filtered.groupby("Country Name")["Completion Rate"].transform(
            lambda x: x.fillna(x.mean()))

        # Daha fazla yƒ±l g√∂stermek i√ßin ilk 10 satƒ±rƒ± alalƒ±m
        df_head = df_filtered.head(10)

        # ƒ∞lk 10 satƒ±rƒ± yazdƒ±r
        print(df_head.to_string(index=False))

        # Global analiz i√ßin t√ºm √ºlkelerin ortalamasƒ±nƒ± alalƒ±m
        df_global = df_filtered.groupby("Year")["Completion Rate"].mean().reset_index()

        # 2020'deki d√º≈ü√º≈ü√º d√ºzeltmek i√ßin 2021 sonrasƒ± i√ßin bir toparlanma trendi ekleyelim
        df_global.loc[df_global["Year"] == pd.Timestamp("2020-01-01"), "Completion Rate"] = \
        df_global[df_global["Year"] < pd.Timestamp("2020-01-01")]["Completion Rate"].mean()

        # Prophet i√ßin veri √ßer√ßevesini uygun hale getirelim
        df_prophet = df_global.rename(columns={"Year": "ds", "Completion Rate": "y"})

        # Prophet modelini olu≈üturuyoruz
        model = Prophet(
            changepoint_prior_scale=0.08,
            seasonality_mode="additive",
            yearly_seasonality=True
        )

        # Manuel deƒüi≈üim noktalarƒ± ekleyelim (Covid-19 etkisini azaltacak ≈üekilde)
        model.add_seasonality(name='yearly', period=1, fourier_order=10)  # Yƒ±llƒ±k etki ekleyelim
        model.changepoints = df_prophet[(df_prophet['ds'] >= '2005-01-01') & (df_prophet['ds'] <= '2020-01-01')]['ds']

        # Modeli eƒüitelim
        model.fit(df_prophet)

        # Gelecek 5 yƒ±l i√ßin tahmin yapalƒ±m (2021-2025)
        future = model.make_future_dataframe(periods=5, freq='Y')
        future['cap'] = 100  # Tahminin √ºst sƒ±nƒ±rƒ±nƒ± belirleyelim
        future['floor'] = df_prophet['y'].min()
        forecast = model.predict(future)

        # G√∂rselle≈ütirme
        plt.figure(figsize=(12, 6))
        plt.plot(df_prophet["ds"], df_prophet["y"], label="Ger√ßek Deƒüerler", marker="o", color="blue")
        plt.plot(forecast["ds"], forecast["yhat"], label="Tahmin (Prophet)", linestyle="dashed", marker="o",
                 color="red")
        plt.fill_between(forecast["ds"], forecast["yhat_lower"], forecast["yhat_upper"], alpha=0.1,
                         label="G√ºven Aralƒ±ƒüƒ±")
        plt.xlabel("Yƒ±l")
        plt.ylabel("Tamamlama Oranƒ± (%)")
        plt.title("Global ƒ∞lk√∂ƒüretim Tamamlama Oranƒ± Tahmini (Prophet) (2021-2025)")
        plt.legend()
        plt.grid(True)
        plt.show()
```

‚Üí 2000-2019 arasƒ±nda istikrarlƒ± bir artƒ±≈ü g√∂zlemlendi ve oran %95 seviyelerine ula≈ütƒ±.

```python
import pandas as pd

# Dosya Yolunu Belirleme
file_path = r"Data\share-of-the-world-population-with-at-least-basic-education.csv"

# Veri Setini Okuma
df = pd.read_csv(file_path, encoding="ISO-8859-1")

#  Gerekli S√ºtunlarƒ± Se√ßme ve Temizlik
df = df.rename(columns={"Entity": "Country", "Year": "Year",
                        "Share of population with no formal education, 1820-2020": "No_Education_Share"})

#  Sadece 2020 Yƒ±lƒ± Verisini Filtreleme
df = df[df["Year"] == 2020]

#  Gerekli S√ºtunlarƒ± Se√ß ve Eksik Deƒüerleri Temizle
df = df[["Country", "Year", "No_Education_Share"]].dropna()

#  Son 10 Satƒ±rƒ± Yazdƒ±r
print(df.tail(10).to_string(index=False))


```

```python
from sklearn.cluster import KMeans

# K-Means K√ºmeleme (3 K√ºme)
X = df[["No_Education_Share"]].values
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(X)

# K√úMELEME SONU√áLARINI CSV'YE KAYDET
df.to_csv("clustering_results.csv", index=False)

print("K√ºmeleme ba≈üarƒ±yla tamamlandƒ±!")
```

```python
# K√ºme 1 ƒ∞√ßindeki √úlkeleri Se√ß
df_cluster_1 = df[df["Cluster"] == 1].copy()

# K√ºme 1‚Äôi 4 Alt K√ºmeye Ayƒ±r
X_sub = df_cluster_1[["No_Education_Share"]].values
kmeans_sub = KMeans(n_clusters=4, random_state=42, n_init=10)
df_cluster_1["Sub_Cluster"] = kmeans_sub.fit_predict(X_sub)

# YENƒ∞ K√úMELERƒ∞ CSV‚ÄôYE KAYDET
df_cluster_1.to_csv("cluster_1_subgroups.csv", index=False)

print(" K√ºme 1, 4 alt k√ºmeye b√∂l√ºnd√º!")
```
